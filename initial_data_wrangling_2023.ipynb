```python
import pandas as pd
import numpy as np
cc = pd.read_excel('STA130 Change Consent Decision(1-273).xlsx')
post = pd.read_excel('STA130 Post-survey (Fall 2023)(1-258).xlsx')
pre = pd.read_excel('STA130 Pre-survey (Fall 2023)(1-533).xlsx')
pre
```

```python
pre['participation_indicator'] = pre['After reading the participant information on Quercus, please select one of the following:'].str.contains('Yes')
pre
```

```python
# now for cc
idx = cc.groupby(['Name', 'Email'])['Completion time'].idxmax()
cleaned_cc = cc.loc[idx]
cleaned_cc
# if name and email are the same we keep the row with the latest completion time!
```

```python
cleaned_cc['participation_indicator'] = cleaned_cc['After reading the participant information on Quercus, please select one of the following:'].str.contains('Yes')
cleaned_cc
# create indicator 
```

```python
final_cc = cleaned_cc[['Email', 'Name', 'participation_indicator']]
final_cc
# we only need these 3 columns 
```

```python
# update pre with indicator from final_cc
final_pre = pd.merge(pre, final_cc, on=['Email', 'Name'], how='left', suffixes=('', '_new'))
final_pre['participation_indicator'] = np.where(final_pre['participation_indicator_new'].notna(), final_pre['participation_indicator_new'], final_pre['participation_indicator'])
final_pre.drop(columns=['participation_indicator_new','Last modified time','Completion time', 'Start time'], inplace=True)
final_pre = final_pre.add_prefix('pre_')
final_pre.rename(columns={'pre_ID': 'ID'}, inplace=True)
final_pre.rename(columns={'pre_Email': 'Email'}, inplace=True)
final_pre.rename(columns={'pre_Name': 'Name'}, inplace=True)
final_pre
# start/finished time droped.
# pre preix added.
```

```python
post
```

```python
post['Name'].value_counts()
# no duplicates!
```

```python
post.drop(columns=['ID', 'Start time','Completion time', 'Last modified time'], inplace=True)
final_post = post.add_prefix('post_')
final_post.rename(columns={'post_Email': 'Email'}, inplace=True)
final_post.rename(columns={'post_Name': 'Name'}, inplace=True)
# drop start/completion/last modified time
# post prefix added 
final_post
```

```python
# now, merge pre and post 
final_data = pd.merge(final_pre, final_post, on=['Email', 'Name'], how='left')
final_data.drop(columns=['Email','Name'], inplace=True)
final_data 
# mission completed!(I hope there's no mistake x.x)
```

```python
# Export final_data to Excel file
final_data.to_excel('final_data_2023.xlsx', index=False)
```
